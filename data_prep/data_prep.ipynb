```python
%matplotlib inline
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt
import cv2
import os
import random

#Original 1600x1327 images of SR
normal = cv2.imread("sr_normal.png")
colored = cv2.imread("sr_colored.png")

h = normal.shape[0]
w = normal.shape[1]

#Where the full lists of 256x256 slices of SR will be stored
col_path = os.getcwd() + r"\colored"
norm_path = os.getcwd() + r"\normal"
```

```python
#Translating a 256x256 window by 96 pixels in each direction repeatedly for 168 unique slices
num = 1
for i in range (0,int(np.ceil((w-256)/96))):
    for j in range (0,int(np.ceil((h-256)/96))):
        norm = normal[min(96*j,h-256):min(256+96*j,h),min(96*i,w-256):min(256+96*i,w)]
        col = colored[min(96*j,h-256):min(256+96*j,h),min(96*i,w-256):min(256+96*i,w)]
        cv2.imwrite(os.path.join(norm_path,"sr" + str(num) + ".jpg"),norm)
        cv2.imwrite(os.path.join(col_path,"sr" + str(num) + ".jpg"),col)
        num = num + 1
        for k in range (0,3):
            #Rotating to augment data
            norm = cv2.rotate(norm,cv2.ROTATE_90_CLOCKWISE)
            col = cv2.rotate(col,cv2.ROTATE_90_CLOCKWISE)
            cv2.imwrite(os.path.join(norm_path,"sr" + str(num) + ".jpg"),norm)
            cv2.imwrite(os.path.join(col_path,"sr" + str(num) + ".jpg"),col)
            num = num + 1
        print(i,j)
```
