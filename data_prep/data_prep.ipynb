```python
%matplotlib inline
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt
import cv2
import os
import random

#2400x1988 images of SR
normal = cv2.imread("sr_normal_2400.png")
colored = cv2.imread("sr_colored_2400.png")

h = normal.shape[0]
w = normal.shape[1]

#Where the full lists of 384x384 slices of SR will be stored
col_path = os.getcwd() + r"\colored_384"
norm_path = os.getcwd() + r"\normal_384"
```

```python
#Translating a 384x384 window by 144 pixels in each direction repeatedly
num = 1
for i in range (0,int(np.ceil((w-384)/144))):
    for j in range (0,int(np.ceil((h-384)/144))):
        norm = normal[min(144*j,h-384):min(384+144*j,h),min(144*i,w-384):min(384+144*i,w)]
        col = colored[min(144*j,h-384):min(384+144*j,h),min(144*i,w-384):min(384+144*i,w)]
        cv2.imwrite(os.path.join(norm_path,"sr" + str(num) + ".jpg"),norm)
        cv2.imwrite(os.path.join(col_path,"sr" + str(num) + ".jpg"),col)
        num = num + 1
        for k in range (0,3):
            #Rotating to augment data
            norm = cv2.rotate(norm,cv2.ROTATE_90_CLOCKWISE)
            col = cv2.rotate(col,cv2.ROTATE_90_CLOCKWISE)
            cv2.imwrite(os.path.join(norm_path,"sr" + str(num) + ".jpg"),norm)
            cv2.imwrite(os.path.join(col_path,"sr" + str(num) + ".jpg"),col)
            num = num + 1
        print(i,j)
```
